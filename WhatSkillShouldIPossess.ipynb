### 4. Model Data

```python
language = [col for col in df_new if col.startswith('LanguageWorkedWith_')]
language_df = df_new[language]

plt.figure(figsize = (16,10))
sns.heatmap(language_df.corr(), xticklabels=[_[19:] for _ in language], yticklabels=[_[19:] for _ in language]);
```

```python
database = [col for col in df_new if col.startswith('DatabaseWorkedWith_')]
database_df = df_new[database]

plt.figure(figsize = (16,10))
sns.heatmap(database_df.corr(), xticklabels=[_[19:] for _ in database], yticklabels=[_[19:] for _ in database]);
```

```python
platform = [col for col in df_new if col.startswith('PlatformWorkedWith_')]
platform_df = df_new[platform]

plt.figure(figsize = (16,10))
sns.heatmap(platform_df.corr(), xticklabels=[_[19:] for _ in platform], yticklabels=[_[19:] for _ in platform]);
```

```python
framework = [col for col in df_new if col.startswith('FrameworkWorkedWith_')]
framework_df = df_new[framework]

sns.heatmap(framework_df.corr(), xticklabels=[_[20:] for _ in framework], yticklabels=[_[20:] for _ in framework]);
```

```python
ide = [col for col in df_new if col.startswith('IDE_')]
ide_df = df_new[ide]

sns.heatmap(ide_df.corr(), xticklabels=[_[4:] for _ in ide], yticklabels=[_[4:] for _ in ide]);
```

```python
devtype = [col for col in df_new if col.startswith('DevType_')]
language = [col for col in df_new if col.startswith('LanguageWorkedWith_')]

dt_lang_df = df_new[devtype + language]
dt_lang_df = dt_lang_df.dropna()

dt_lang_df.head()
```

### 5. Results

```python
dt_lang_perc = pd.DataFrame(columns=devtype, index=language)

for dt in devtype:
    dt_lang_perc[dt] = dt_lang_df[dt_lang_df[dt] == 1][language].sum()/dt_lang_df[dt_lang_df[dt] == 1].count()

dt_lang_perc.index = dt_lang_perc.index.str.slice(19)

plt.figure(figsize=(20, 20))
for i, dt in enumerate(devtype):
    plt.subplot(4, 5, i+1)
    dt_lang_perc[dt].sort_values()[-10:].plot(kind='barh', title=dt[8:], sharex=True)

plt.tight_layout()
```
